```python
# 1. Use matplotlib to plot the following equation: y = x^2 - x + 2
#  You'll need to write the code that generates the x and y points.
#  Add an anotation for the point 0, 0, the origin.

x = np.array(range(10))
y = x ** 2 - x + 2

plt.plot(x,y)
plt.annotate('origin', xy=(0,0), xytext=(1,15), arrowprops = {})
```

```python
# 2.

x = np.array(range(4,26))
y0 = np.sqrt(x)

plt.plot(x,y0)

plt.grid(True, ls='--')
```

```python
y1 = x ** 3

plt.plot(x,y1)
```

```python
y2 = 2 ** x

plt.plot(x,y2)
```

```python
y3 = 1/(x+1)

plt.plot(x,y3)
```

```python
# 3. Combine the figures you created in the last step into one large figure with 4 subplots.
n_rows = 1
n_cols = 2

plt.subplot(221)
plt.plot(x,y0)
plt.legend(['$y = sqrt(x)$'])

plt.subplot(222)
plt.plot(x,y1)
plt.legend(['$y = x^3$'])

plt.subplot(223)
plt.plot(x,y2)
plt.legend(['$y = 2^x$'])

plt.subplot(224)
plt.plot(x,y3)
plt.legend(['y = 1/(x + 1)'])

plt.suptitle('Equations stuff')

plt.tight_layout() # sets padding or puts a good distance between each plot
```

```python

plt.figure(figsize=(10,8))

plt.plot(x,y0, c = 'blue', alpha = .8) # alpha = opacity
plt.plot(x,y1, c = 'red')
plt.plot(x,y2, c = 'orange')
plt.plot(x,y3, c = 'green')
plt.legend(['$y = sqrt(x)$','$y = x^3$','$y = 2^x$','y = 1/(x + 1)'])

```
