```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
import scipy
```

```python
np.random.seed(123)

n = np.array(range(1, 15))
n
```

```python
y = 0*n + 1     # "O(1)"

y0 = np.log(n)   # "O(log n)" 

y1 = n  # "O(n)"

y2 = n * np.log(n)   # "O(n log n)"

y3 = n ** 2   # "O(n^2)"

y4 = 2 ** n # "O(2^n)"

y5 = scipy.special.factorial(n) # "O(n!)"

y6 = n ** n   # "O(n^n)"
```

```python
plt.figure(figsize=(15, 10))

plt.suptitle('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.plot(n,y)  # "O(1)"

plt.plot(n,y0) # "O(log n)" 

plt.plot(n,y1) # "O(n)"

plt.plot(n,y2) # "O(n log n)"

plt.plot(n,y3) # "O(n^2)"

plt.plot(n,y4) # "O(2^n)"

plt.plot(n,y5) # "O(n!)"

plt.plot(n,y6) # "O(n^n)"

plt.ylim(1,30)

plt.legend(['y=0n+1', 'y=log(n)', 'y=n', 'y=n*log(n)','$y=n^2$','$y=2^n$','y=n!','$y=n^n$'], loc ='upper right')

plt.annotate('y=n', xy=(8,8), xytext=(7,9), arrowprops={})
```
